```python
### fill the null values with the fill na

df['director']=df['director'].fillna(df['director'].mode()[0]) ## mode is only the object
```

```python
df['cast']=df['cast'].fillna(df['cast'].mode()[0])
```

```python
df['country']=df['country'].fillna(df['country'].mode()[0])
```

```python
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
```

```python
df['duration']=df['duration'].fillna(df['duration'].mode()[0])
```

```python
#### firtsly convert the date in seperated from

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.head(3)
```

```python
df['year']=df['date_added'].dt.year
```

```python
df['month']=df['date_added'].dt.month
```

```python
df['day']=df['date_added'].dt.day
```

```python
# filll the fulll value of year,month,daay
df['year']=df['year'].fillna(df['year'].mean())
df['month']=df['month'].fillna(df['month'].mean())
df['day']=df['day'].fillna(df['day'].mean())
```

```python
df.head(4)
```

```python
### convert the data type of the year, month,day

df['year']=df['date_added'].astype(int)
df['month']=df['date_added'].astype(int)
df['day']=df['date_added'].astype(int)
```

```python
df.isnull().sum()
```

```python
### date_added column is delete from the dataset after all process over it.
df.drop('date_added',axis=1,inplace=True)
```

```python
## replace the value of s with whitespace

df['show_id']=df['show_id'].str.replace('s','')
```

```python
df.head(2)
```

```python
### convert the data type of show_id
df['show_id']=df['show_id'].astype(int)
```

```python
## convert the data type of release_year
df['release_year']=df['release_year'].astype(int)
```

```python
def convert_to_minutes(x):
    if pd.isna(x):
        return np.nan

    x = x.lower().strip()

    if 'min' in x:
        return int(x.replace('min', '').strip())

    if 'season' in x:          # case-insensitive, catches both 'season' and 'seasons'
        seasons = int(x.split()[0])
        return seasons * 450   # 450 min per season (≈ 7.5 h)

    return np.nan

df['duration_minutes']=df['duration'].apply(convert_to_minutes)
```

```python
df['duration_minutes']=df['duration_minutes'].fillna(df['duration'].mode()[0])
```

```python
df.drop('year',axis=1,inplace=True)
df.drop('month',axis=1,inplace=True)
df.drop('day',axis=1,inplace=True)
```

```python
df.drop('duration',axis=1,inplace=True)
```

```python
#####feature engneering
features=["type","director","cast","country","listed_in","description","release_year","duration_minutes"]
df_enc=df[features].fillna("")
```

###convert the text to numerical vector



```python

from sklearn.feature_extraction.text import TfidfVectorizer
text_features=(
    df_enc['director']+''+
    df_enc['cast']+''+
    df_enc['country']+''+
    df_enc['listed_in']+''+
    df_enc['description']
)

tfidf=TfidfVectorizer(stop_words="english",max_features=5000)
text_encoded=tfidf.fit_transform(text_features)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
type_encoded=encoder.fit_transform(df_enc[['type']])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

numeric_encoded=scaler.fit_transform(df_enc[['release_year','duration_minutes']])
```

```python
from scipy.sparse import hstack
final_features=hstack([text_encoded,type_encoded,numeric_encoded])
```

```python
final_features
```

```python
final_features=final_features.tocsr()
```
